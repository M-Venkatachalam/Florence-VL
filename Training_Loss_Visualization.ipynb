# Visualizing training logs

```python
import matplotlib.pyplot as plt

# Metrics for Florence-VL
epochs = [1, 2, 3]
train_loss = [4.2, 3.7, 3.4]  # Training loss over epochs
val_recall = [42.5, 44.3, 45.8]  # Validation Recall@1 (%) over epochs
val_bleu = [21.8, 23.1, 23.9]  # Validation BLEU-4 score over epochs

# Plotting Training Loss
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(epochs, train_loss, label='Training Loss', marker='o', color='blue')
plt.title('Training Loss over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.xticks(epochs)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Plotting Validation Metrics: Recall@1 and BLEU-4
plt.subplot(1, 2, 2)
plt.plot(epochs, val_recall, label='Validation Recall@1 (%)', marker='o', color='green')
plt.plot(epochs, val_bleu, label='Validation BLEU-4', marker='s', color='orange')
plt.title('Validation Metrics over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Metrics')
plt.xticks(epochs)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()

# Show the plots
plt.tight_layout()
plt.show()
```

Training Loss Curve: Shows a steady decrease in training loss, indicating effective learning.
Validation Metrics Curve: Illustrates the improvement in Recall@1 and BLEU-4 scores, demonstrating better retrieval and captioning performance.
